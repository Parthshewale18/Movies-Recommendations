### Cleaning the Data

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['tags'] = movies['keywords']+movies['genres']+movies['cast']+movies['crew']
movies['tags'][7]
```

```python
movies = movies[['movie_id','title','overview','tags']]
movies.head(3)
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

### Traning Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(movies['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
def get_recommendations(title,cosine_sim=cosine_sim):
    idx = movies[movies['title']==title].index[0] # Find the index of the movie in the DataFrame
    sim_scores = list(enumerate(cosine_sim[idx])) # Get all similarity scores for this movie
    sim_scores = sorted(sim_scores,key=lambda x:x[1], reverse=True) # Sort movies by similarity score (highest first)
    sim_scores = sim_scores[1:11] # Get Top 10 similar movies
    movies_indices = [i[0] for i in sim_scores]# Get the indices of those top movies
    return movies['title'].iloc[movies_indices]
```
